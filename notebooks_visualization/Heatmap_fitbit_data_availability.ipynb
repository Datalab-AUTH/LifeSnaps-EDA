```python
# Loading the dataframes with hourly granularity

file = open(r'data\users_distance_hourly','rb')
distance = pickle.load(file)
file.close()
distance.reset_index(inplace=True)

distance['date'] = pd.to_datetime(distance.date, format='%m/%d/%y %H:%M:%S') 
distance['distance'] = pd.to_numeric(distance['distance'])
distance = distance.astype({"id": str})
distance = distance.sort_values('date')

# Splitting timestamp column into two separate date and time columns
distance['Dates'] = pd.to_datetime(distance['date']).dt.date
distance['Time'] = pd.to_datetime(distance['date']).dt.time

distance['Dates'] = pd.to_datetime(distance.Dates, format='%Y/%m/%d') 
#distance1 = distance[(distance['Dates'] >= "2021-05-24") & (distance['Dates'] <= "2021-07-26")] #Round1
#distance2 = distance[(distance['Dates'] >= "2021-11-15") & (distance['Dates'] <= "2022-01-17")] #Round2

#distance=distance1.append(distance2)
distance=distance[['Dates','Time','id','distance']]
distance
```

```python
# Loading the dataframes with hourly granularity

file = open(r'data\users_estimated_oxygen_variation_hourly','rb')
estimated_oxygen_variation = pickle.load(file)
file.close()

estimated_oxygen_variation['date'] = pd.to_datetime(estimated_oxygen_variation.date, format='%m/%d/%y %H:%M:%S') 
estimated_oxygen_variation['estimated_oxygen_variation'] = pd.to_numeric(estimated_oxygen_variation['estimated_oxygen_variation'])
estimated_oxygen_variation = estimated_oxygen_variation.astype({"id": str})
estimated_oxygen_variation = estimated_oxygen_variation.sort_values('date')

# Splitting timestamp column into two separate date and time columns
estimated_oxygen_variation['Dates'] = pd.to_datetime(estimated_oxygen_variation['date']).dt.date
estimated_oxygen_variation['Time'] = pd.to_datetime(estimated_oxygen_variation['date']).dt.time

estimated_oxygen_variation['Dates'] = pd.to_datetime(estimated_oxygen_variation.Dates, format='%Y/%m/%d') 
#estimated_oxygen_variation1 = estimated_oxygen_variation[(estimated_oxygen_variation['Dates'] >= "2021-05-24") & (estimated_oxygen_variation['Dates'] <= "2021-07-26")] #Round1
#estimated_oxygen_variation2 = estimated_oxygen_variation[(estimated_oxygen_variation['Dates'] >= "2021-11-15") & (estimated_oxygen_variation['Dates'] <= "2022-01-17")] #Round2

#estimated_oxygen_variation=estimated_oxygen_variation1.append(estimated_oxygen_variation2)
estimated_oxygen_variation=estimated_oxygen_variation[['Dates','Time','id','estimated_oxygen_variation']]
estimated_oxygen_variation
```

```python
# Loading the dataframes with hourly granularity

file = open(r'data\users_steps_hourly','rb')
steps = pickle.load(file)
file.close()

steps['date'] = pd.to_datetime(steps.date, format='%m/%d/%y %H:%M:%S') 
steps['steps'] = pd.to_numeric(steps['steps'])
steps = steps.astype({"id": str})
steps = steps.sort_values('date')

# Splitting timestamp column into two separate date and time columns
steps['Dates'] = pd.to_datetime(steps['date']).dt.date
steps['Time'] = pd.to_datetime(steps['date']).dt.time

steps['Dates'] = pd.to_datetime(steps.Dates, format='%Y/%m/%d') 
#steps1 = steps[(steps['Dates'] >= "2021-05-24") & (steps['Dates'] <= "2021-07-26")] #Round1
#steps2 = steps[(steps['Dates'] >= "2021-11-15") & (steps['Dates'] <= "2022-01-17")] #Round2

#steps=steps1.append(steps2)
steps=steps[['Dates','Time','id','steps']]
steps
```

We removed calories as fitbit computes calories even while not wearing the fitbit.

```python
#df = pd.concat([steps, distance, calories, estimated_oxygen_variation])
#df = pd.concat([steps, distance, calories])
df = pd.concat([steps, distance, estimated_oxygen_variation])
df
```

```python
# Replace NaNs with 0 for the whole dataframe
df=df.fillna(0)

# Replace non-zero values with 1
#df=df.astype(bool).astype(int)
df['steps'] = df['steps'].where(df['steps'] == 0, 1)
df['distance'] = df['distance'].where(df['distance'] == 0, 1)
#df['calories'] = df['calories'].where(df['calories'] == 0, 1)
df['estimated_oxygen_variation'] = df['estimated_oxygen_variation'].where(df['estimated_oxygen_variation'] == 0, 1)
#df['fitbit_data'] = df['steps'] + df['distance'] + df['calories'] + df['estimated_oxygen_variation']
df['fitbit_data'] = df['steps'] + df['distance'] + df['estimated_oxygen_variation']
df=df[['Dates','Time','id','fitbit_data']]

# Delete all zero entries
df=df[(df != 0).all(1)]

df
```

```python
df=df[['Dates','Time','id']]
df
#df.head(100)
```

```python
df.to_pickle(r'data\fitbit_data_availability' )
```

```python
df=df.groupby(['Dates','Time']).agg('id').nunique()
df=pd.DataFrame(df)
df.reset_index(inplace=True)
df.rename(columns = {'id':'NumberOfDistictUsers'}, inplace = True)
#df.head(500)
df
```

```python
df.to_pickle(r'data\users_data_fitbit_availability_hourly' )
```

```python
# Separating the two Rounds
df1 = df[(df['Dates'] >= "2021-05-24") & (df['Dates'] <= "2021-07-26")] #Round1
df2 = df[(df['Dates'] >= "2021-11-15") & (df['Dates'] <= "2022-01-17")] #Round2
```

```python
df1['Dates'] = pd.to_datetime(df1.Dates, format='%Y/%m/%d') 
df1['Dates'] = pd.to_datetime(df1['Dates']).dt.date
```

```python
#piv = df.pivot_table(index='Time',columns='Dates',values='NumberOfDistictUsers')
piv1 = pd.pivot_table(df1, values="NumberOfDistictUsers",index=["Time"], columns=["Dates"], fill_value=0)
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.heatmap(piv1,cmap=cmap)
xticks = plt.gca().xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i % 5 != 0:
        xticks[i].set_visible(False)
plt.setp( ax.xaxis.get_majorticklabels(), rotation=0 )
plt.tight_layout()
plt.tight_layout()
ax.yaxis.set_major_locator(md.HourLocator(interval = 100))
ax.yaxis.set_major_formatter(md.DateFormatter('%H:%M:%S'))
ax.set_rasterization_zorder(0)
plt.xlabel('Study date', fontsize=45)
plt.ylabel('Time of the day', fontsize=45)
plt.text(74, 21, 'Number of distinct users', rotation=270, fontsize=35)
plt.title('Fitbit Data Availability: Round 1', fontsize=45)
plt.savefig(r"visualizations\Round1_fitbit_data_availability_heatmap.eps", bbox_inches='tight') 
plt.savefig(r"visualizations\Round1_fitbit_data_availability_heatmap.png", bbox_inches='tight') 
plt.show()
```
